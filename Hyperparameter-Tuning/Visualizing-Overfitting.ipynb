```python
import pandas as pd
import numpy as np
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
alt.renderers.enable('notebook')
alt.themes.enable('opaque')
```

```python
depth_10 = pd.read_csv('data/depth_10.csv')
depth_10.head()
```

```python
c1 = alt.Chart(depth_10).mark_line(color='blue').encode(
    alt.X('iteration:N'),
    alt.Y('training:Q', scale=alt.Scale(zero=False)),
)

c2 = alt.Chart(depth_10).mark_line(color='red').encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
)

c1 + c2
```
