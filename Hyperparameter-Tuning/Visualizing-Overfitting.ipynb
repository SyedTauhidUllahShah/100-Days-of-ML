```python
import pandas as pd
import numpy as np
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
alt.renderers.enable('notebook')
alt.themes.enable('opaque')
```

```python
depth_10 = pd.read_csv('data/depth_10.csv')
depth_10.head()
```

```python
depth_9 = pd.read_csv('data/depth_9.csv')
depth_9.head()
```

```python
depth_8 = pd.read_csv('data/depth_8.csv')
depth_8.head()
```

```python
c1 = alt.Chart(depth_10).mark_line(color='blue', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('training:Q', scale=alt.Scale(zero=False)),
    tooltip = ['training']
)

c2 = alt.Chart(depth_10).mark_line(color='red', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
    tooltip = ['test']
)

(c1 + c2).interactive()
```

```python
c1 = alt.Chart(depth_9).mark_line(color='blue', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('training:Q', scale=alt.Scale(zero=False)),
    tooltip = ['training']
)

c2 = alt.Chart(depth_9).mark_line(color='red', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
    tooltip = ['test']
)

(c1 + c2).interactive()
```

```python
c1 = alt.Chart(depth_8).mark_line(color='blue', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('training:Q', scale=alt.Scale(zero=False)),
    tooltip = ['training']
)

c2 = alt.Chart(depth_8).mark_line(color='red', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
    tooltip = ['test']
)

(c1 + c2).interactive()
```

```python
c1 = alt.Chart(depth_8).mark_line(color='blue', interpolate='step-after').encode(
    alt.X('iteration:O'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
)

c1
```

```python
c1 = alt.Chart(depth_10).mark_line(color='blue', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
    tooltip = ['test']
)

c2 = alt.Chart(depth_9).mark_line(color='red', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
    tooltip = ['test']
)

c3 = alt.Chart(depth_8).mark_line(color='green', point = True).encode(
    alt.X('iteration:N'),
    alt.Y('test:Q', scale=alt.Scale(zero=False)),
    tooltip = ['test']
)

(c1 + c2 + c3).interactive()
```
